#1. 1) Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800). Найдите ее среднее значение и дисперсию.

Описание решения:
т.к равномерное распределения на промежутке (a, b), то среднее значение (μ) вычисляется по формуле: μ = (a + b) / 2


```python
import math
# иисходное дано
a = 200  
b = 800  

# считаем среднее значение (mean)
mean = (a + b) / 2
print("среднее значение:", mean)


```

а дисперсия (что равна σ^2) равномерного распределения на промежутке (a, b) вычисляется по формуле:
σ^2 = (b - a)^2 / 12

```python
import math
# иисходное дано
a = 200  
b = 800 

# дисперсию считаем (variance)
variance = ((b - a) ** 2) / 12
print("дисперсия:", variance)
```

#5.На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?


Описание и решение.

Для определения на сколько сигм (средних квадратичных отклонений) отклоняется значение от среднего, мы используем формулу стандартизации:

Z = (X - M) / sqrt(D), где: Z - стандартизованное значение, X - значение, от которого мы хотим определить отклонение, M - математическое ожидание (среднее значение), D - дисперсия.

В данной задаче:
X = 190 см,
M = 178 см,
D = 25 кв.см.

подстпавляем имеющиеся значения в формулу, получаем: 
Z = (190 - 178) / sqrt(25) = 12 / 5 = 2.4

значит человек с ростом 190 см  отклоняется на 2.4 сигмы (средних квадратичных отклонений) от математического ожидания роста в этой популяции.
