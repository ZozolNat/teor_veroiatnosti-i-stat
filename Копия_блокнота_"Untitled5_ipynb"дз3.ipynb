#1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

Описание решения:

сначала среднее арифметическое (сложить и поделить на колличество )

```python
import math

# исходные данные - значения зарплат
salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# среднее арифметическое
mean = sum(salaries) / len(salaries)
print(f"Среднее арифметическое: {mean}")
```

далее среднее квадратическое отклонение (1.из каждого значения вычесть среднее арифметическое и возвести в квадрат, т.е квадрат отклонения этого значения; 2. найти среднее значение квадратов отклонения, т.е сложить все значения квадратов отклонения и разделить на их колличество ; 3. извлечь квадратный корень из среднего значения квадратов отклонения)

```python
# среднее квадратичное отклонение
variance = sum([((x - mean) ** 2) for x in salaries]) / len(salaries)
standard_deviation = math.sqrt(variance)
print(f"Среднее квадратичное отклонение: {standard_deviation}")
```

смещенная оценка дисперсий для данной выборки (суммирование квадратов отклонений каждого значения от среднего арифметического выборки, и деления этой суммы на количество значений в выборке, .т.е пункт 1 и 2 в предыдущем действии)

```python
# смещенная оценка дисперсий, предыдущее без извлечения квадратного корня
variance = sum([((x - mean) ** 2) for x in salaries]) / len(salaries)
(f"смещенная оценка дисперсий: {variance}")
```

несмещенная оценка дисперсий для данной выборки (тоже самое, только из колличества элементов вычитаем 1)

```python
# несмещенная оценка дисперсий, предыдущее без извлечения квадратного корня
variance = sum([((x - mean) ** 2) for x in salaries]) / (len(salaries)-1)
(f"несмещенная оценка дисперсий: {variance}")
```

B - из 1 ящика выщили 1 белый мяч, и из 2 ящика вытащили 2 бел мяча;
P(B) = C(5,1)*C(3,1)/C(8,2) * C(5,2)*C(7,2)/C(12,4) ≈ 0.2045

C - из 1 ящика 0 белых мячей,  из 2 ящика вытащили 3 белых мяча.
P(C) = C(3,2)/C(8,2) * C(5,3)*C(7,1)/C(12,4) ≈ 0.0045


Далее складываем эти вероятности
P = P(A) + P(B) + P(C) ≈ 0.2257

#3.На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.


Описание и решение.




Пусть A1, A2 и A3 - события "выстрел произведен первым, вторым, третьим спортсменом соответственно", а B - событие "было попадание в мишень". Тогда:
Принимаем , что 1/3 вероятность, что выстрел был сделан или 1 или 2 или 3 стрелком, т.е P(A1) = 1/3, P(A2) = 1/3, P(A3) = 1/3 

Далее принимаем, 

P(B|A1) = 0.9 - вероятность попадания при условии, что выстрела сделанного 1 спортсменом, P(B|A2) = 0.8 - 2ым спортсменом, 
P(B|A3) = 0.6 - 3ьим спортсменом.

Используем формулу Байеса: P(A|B) = P(B|A)*P(A)/P(B), где A и B - события.

Считаем вероятнрсть, что этот выстрел, который попал в мишень был сделан 1ым спортсменом , по формуле:

P(A1|B) = P(B|A1)*P(A1)/P(B) = P(B|A1)*P(A1)/(P(B|A1)*P(A1) + P(B|A2)*P(A2) + P(B|A3)*P(A3))
и так для каждого спортсмена соответвенно 2 ого и 3 его.

P(A1|B) = 0.9*(1/3)/(0.9*(1/3) + 0.8*(1/3) + 0.6*(1/3)) = 0.3913
P(A2|B) = 0.8*(1/3)/(0.9*(1/3) + 0.8*(1/3) + 0.6*(1/3)) = 0.3478
P(A3|B) = 0.6*(1/3)/(0.9*(1/3) + 0.8*(1/3) + 0.6*(1/3)) = 0.2609

в процентах 39,1%, 34,8%, 26,1% соответвенно

Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

За D принммаем вероятности события и используем формулу полной вероятности P(D) = P(D|A)P(A) + P(D|B)P(B) + P(D|C)P(C):
