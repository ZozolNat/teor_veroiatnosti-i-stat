#1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Описание решения а).

Мы знаем, что в колоде 13 крестей.
P = (количество способов вытянуть 4 карты крести) / (количество способов вытянуть 4 карты из всей колоды).
Количество способов вытянуть 4 карты крести можно вычислить следующим образом:
Выбираем 4 карты из 13 крестей: C(13, 4)
Количество способов выбрать 4 карты из колоды составляет C(52, 4)
Тогда вероятность вытянуть 4 карты крести будет равна:
P = C(13, 4) / C(52, 4)


пояснение- С(13, 4), это С вверху 4 внизу 13.


```python
import math

# количество способов выбрать 4 карты из 13 крестей
k = math.comb(13, 4)

# количество способов выбрать 4 карты из всей колоды
n = math.comb(52, 4)

# вероятность вытянуть 4 карты крести
p = k / n

print("Вероятность вытянуть 4 карты крести:", p)
```

Описание решения б).

Используем принципом дополнения: найдем вероятность того, что среди 4-х карт не будет ни одного туза и вычтем ее из 1.

Количество способов вытянуть 4 карты без тузов можно вычислить: C(48, 4). (тоесть 48 карт это не тузы плюс 4 карты это тузы.)

А вероятность того, что среди 4-х карт нет ни одного туза будет равна отношение  колличества способов вытащить без туза см ввверху к общему количеству.общее количество карт 52 :

P = C(48, 4) / C(52, 4) 

далее из единицы вычитаем вероятность , что среди вытащенных карт нет туза и получаем.
Вероятность того, что хотя бы один туз есть среди 4-х карт, будет равна:

P = 1 - C(48, 4) / C(52, 4)

```python
import math

# количество способов выбрать 4 карты из 48 карт без тузов
k = math.comb(48, 4)

# количество способов выбрать 4 карты из всей колоды
n = math.comb(52, 4)

# вероятность того, что среди 4-х карт нет ни одного туза
p_no_ace = k / n

# вероятность того, что хотя бы один туз есть среди 4-х карт
p_at_least_one_ace = 1 - p_no_ace

print("Вероятность того, что хотя бы один туз есть среди 4-х карт:", p_at_least_one_ace)
```

#3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

Решение и описание.
Сначало считаем число сочетаний, что  из 9 окрашенных достаем по 3 окрашенных, это представим, как C(9,3) 

используем формулу: C(n,k) = n! / (k! * (n-k)!)

и получаем C(9,3) = 9! / (3! * (9-3)!) = 9! / (3! * 6!) = (9 * 8 * 7) / (3 * 2 * 1) = 84

Далее считаем всего возможные сочетания вытаскивая 3 детали из общего количества C(15,3) 
= 15! / (3! * (15-3)!) =(15 * 14 * 13) / (3 * 2 * 1) = 455 сочетаний при вытаскивании 3 деталей из 15 деталей.

Т.о., вероятность извлечения трех окрашенных деталей равна отношению числа сочетаний, содержащих только окрашенные детали (84), к общему числу сочетаний (455):

P = 84 / 455 = 0.185 -все вытащенные детали будут окрашены



Описание и решение.

Используем формулу для вычисления вероятности совместного наступления событий:

P(A и B) = P(A) * P(B | A)

 P(A) - вероятность наступления события A, P(B | A) - вероятность наступления события B, только в случае уже наступишнего события A.

примем за A - первый билет является выигрышным, за B - второй билет тоже является выигрышным.
Берем два положительных исхода из общего, т.е. выиграть с первой попытки купив два билета это 2/100 = 0.02 (так как всего 100 билетов и 2 из них выигрышные). Если первый билет выигрышный, то остается 99 билетов, включая 1 выигрышный из них, поэтому вероятность выиграть второй билет при условии, что первый был выигрышным, равна 1/99 = 0.01.

Таким образом, вероятность того, что оба купленных билета окажутся выигрышными, считаем по формуле:

P(A и B) = P(A) * P(B | A) = 0.02 * 0.01 = 0.000202
